```python
import cv2
```

```python
# Read an Image, by default its BGR color space
img = cv2.imread('cmonbrug.jpg')

# Reading as RGB
newImg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(img)
plt.show()
plt.imshow(newImg)
plt.show()
```
